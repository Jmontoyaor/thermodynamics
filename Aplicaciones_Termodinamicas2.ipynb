```python
%%writefile A-Bomba.py
# -*- coding: utf-8 -*-
import streamlit as st
from iapws import IAPWS97


custom_css = """

<style>
    .stApp {
        background-color: #0e1a40;
        color: #E0E0E0;
        font-family: 'Courier New', monospace;
    }

    h1, h2, h3 {
        color: #00BFFF;
    }

    .main .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
        background-color: #1B1D2B;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] {
        background-color: #222f5b;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] * {
        color: #c6e2ff !important;
    }

    .resultado-final {
        color: #FFD700;
        background-color: #2c3e50;
        border: 1px solid #FFD700;
        border-radius: 8px;
        padding: 1rem;
        text-align: center;
        font-size: 1.1rem;
    }

    .nota-info {
        color: #98FB98;
    }


    /* Estilo para el texto de la nota */
    .nota-info {
        color: #98FB98; /* Color Verde P√°lido para las letras */
    }


</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)

# -----------------------------------------------------------------------------
# CONFIGURACI√ìN DE LA P√ÅGINA DE STREAMLIT
# -----------------------------------------------------------------------------
st.set_page_config(
    page_title="C√°lculo de Bomba",
    layout="centered",
    initial_sidebar_state="expanded"
)

# -----------------------------------------------------------------------------
# FUNCI√ìN DE C√ÅLCULO (sin cambios)
# -----------------------------------------------------------------------------
def calcular_bomba(
    caudal_volumetrico_m3_s: float,
    presion_entrada_kpa: float,
    presion_salida_mpa: float
) -> dict:
    """
    Calcula el trabajo, la potencia y la entalp√≠a de salida para una bomba.
    """
    # Conversi√≥n de Unidades
    presion_entrada_mpa = presion_entrada_kpa / 1000
    presion_salida_kpa = presion_salida_mpa * 1000

    # PASO 1: Propiedades a la entrada (L√≠quido saturado)
    try:
        estado_entrada = IAPWS97(P=presion_entrada_mpa, x=0)
        v_especifico = estado_entrada.v
        h1_entalpia = estado_entrada.h
    except Exception as e:
        st.error(f"Error al obtener propiedades del agua: {e}")
        st.stop()

    # PASO 2: Flujo m√°sico
    flujo_masico_kgs = caudal_volumetrico_m3_s / v_especifico

    # PASO 3: Trabajo espec√≠fico
    delta_presion_kpa = presion_salida_kpa - presion_entrada_kpa
    trabajo_especifico_kj_kg = v_especifico * delta_presion_kpa

    # PASO 4: Potencia de la bomba
    potencia_kw = flujo_masico_kgs * trabajo_especifico_kj_kg

    # PASO 5: Entalp√≠a de salida
    h2_entalpia = h1_entalpia + trabajo_especifico_kj_kg

    # Empaquetar resultados
    return {
        "entradas": {
            "caudal_volumetrico_m3_s": caudal_volumetrico_m3_s,
            "presion_entrada_kpa": presion_entrada_kpa,
            "presion_salida_mpa": presion_salida_mpa
        },
        "propiedades_entrada": {
            "volumen_especifico_m3_kg": v_especifico,
            "entalpia_h1_kj_kg": h1_entalpia
        },
        "calculos_intermedios": {
            "flujo_masico_kgs": flujo_masico_kgs,
        },
        "resultados_finales": {
            "trabajo_especifico_kj_kg": trabajo_especifico_kj_kg,
            "potencia_requerida_kw": potencia_kw,
            "entalpia_h2_kj_kg": h2_entalpia
        }
    }

# -----------------------------------------------------------------------------
# INTERFAZ DE USUARIO CON STREAMLIT
# -----------------------------------------------------------------------------

# --- T√≠tulo de la Aplicaci√≥n ---
# =========================
# üöÄ Interfaz Streamlit
# =========================
col_img, col_title = st.columns([0.2, 1])
with col_img:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Bomba.png", width=200)
with col_title:
    st.title("Bomba: An√°lisis interactivo de Bomba")
st.markdown("#### Esta herramienta calcula la potencia y entalp√≠a para una bomba que maneja agua como l√≠quido incompresible.")
st.write("---")



# =========================
# üìò Teor√≠a
# =========================
with st.expander("üìò Fundamentos Termodin√°micos de Bombas Centr√≠fugas"):
    st.markdown(r"""
Las **bombas** son dispositivos mec√°nicos dise√±ados para **incrementar la presi√≥n de un l√≠quido**, transport√°ndolo de una regi√≥n de menor presi√≥n a otra de mayor presi√≥n. Son ampliamente usadas en sistemas de suministro de agua, plantas termoel√©ctricas, procesos industriales y sistemas de calefacci√≥n y refrigeraci√≥n.

A diferencia de toberas y difusores, **las bombas s√≠ realizan trabajo mec√°nico** sobre el fluido. En t√©rminos de la **Primera Ley de la Termodin√°mica para sistemas abiertos (volumen de control)**, la potencia de bombeo se calcula como:

$$
\dot{W}_{bomba} = \dot{m} (h_2 - h_1)
$$

donde:
- $\dot{m}$ es el flujo m√°sico de l√≠quido
- $h_1$ y $h_2$ son las entalp√≠as espec√≠fica a la entrada y salida de la bomba, respectivamente

En l√≠quidos **incompresibles**, la variaci√≥n de entalp√≠a puede expresarse como una diferencia de presi√≥n:

$$
h_2 - h_1 = v (P_2 - P_1)
$$

donde:
- $v$ es el **volumen espec√≠fico**, pr√°cticamente constante
- $P_1$ y $P_2$ son las presiones a la entrada y salida

De esta forma, la potencia requerida por la bomba es:

$$
\dot{W}_{bomba} = \dot{m} v (P_2 - P_1)
$$

En la pr√°ctica, se debe considerar la **eficiencia de la bomba**, que relaciona la potencia hidr√°ulica √∫til con la potencia suministrada por el motor.
""")



# =========================
# üß™ Ejercicio y Figura
# =========================
col1, col2 = st.columns(2)
with col1:
  with st.expander("üìö Ejercicio Propuesto ‚Äì An√°lisis de Bomba"):
    st.markdown("""
**Primera Ley de la Termodin√°mica ‚Äì Sistemas Abiertos**

**BOMBAS**

Entra un flujo volum√©trico de **agua** igual a **0.015645 m¬≥/s** con una **presi√≥n de 100 kPa** a una bomba, y sale con una **presi√≥n de 2.5 MPa**.
Calcular:

- a) La potencia de la bomba por unidad de masa \([kJ/kg]\)
- b) La potencia de la bomba en kW
- c) La entalp√≠a del agua a la salida de la bomba

---

**Fuente:** *Ejercicio tomado y adaptado de **LaMejorAsesor√≠aEducativa ‚Äì YouTube***.
""")

with col2:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/partes-de-una-bomba-centrifuga.jpg",
             caption="**Partes de una bomba centr√≠fuga .\\n\\nFuente: Seguas ‚Äì Diagrama de bomba centr√≠fuga (segua‚Äãs.com).")
    st.markdown("### Desarrollo visual")
    st.video("https://youtu.be/yqasAH1LFOw?si=_TbbgdZrU_uff-JP")
st.markdown("---")



# --- Barra Lateral con Controles de Entrada ---
st.sidebar.header("Par√°metros de Entrada")

v_dot = st.sidebar.slider(
    'Caudal Volum√©trico (VÃá) [m¬≥/s]',
    min_value=0.001,
    max_value=0.1,
    value=0.0123,
    step=0.0001,
    format="%.4f"
)

p1 = st.sidebar.slider(
    'Presi√≥n de Entrada (P‚ÇÅ) [kPa]',
    min_value=10.0,
    max_value=1000.0,
    value=159.0,
    step=1.0,
    format="%.1f"
)

p2 = st.sidebar.slider(
    'Presi√≥n de Salida (P‚ÇÇ) [MPa]',
    min_value=0.5,
    max_value=15.0,
    value=3.5,
    step=0.1,
    format="%.2f"
)

# --- Realizar el c√°lculo con los valores de la UI ---
calculo = calcular_bomba(
    caudal_volumetrico_m3_s=v_dot,
    presion_entrada_kpa=p1,
    presion_salida_mpa=p2
)

# --- Extracci√≥n de resultados para mostrarlos ---
entradas = calculo["entradas"]
props = calculo["propiedades_entrada"]
intermedios = calculo["calculos_intermedios"]
finales = calculo["resultados_finales"]

# -----------------------------------------------------------------------------
# PRESENTACI√ìN DE RESULTADOS EN EL √ÅREA PRINCIPAL
# -----------------------------------------------------------------------------

st.header("Resultados del C√°lculo")
st.markdown("---")

# --- Resumen Final en la parte superior ---
st.subheader("Resumen Final")
col1, col2, col3 = st.columns(3)

with col1:
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Potencia Requerida:</strong><br>
            {finales['potencia_requerida_kw']:.2f} kW
        </div>
        """, unsafe_allow_html=True
    )

with col2:
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Trabajo Espec√≠fico:</strong><br>
            {finales['trabajo_especifico_kj_kg']:.4f} kJ/kg
        </div>
        """, unsafe_allow_html=True
    )

with col3:
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Entalp√≠a de Salida (h‚ÇÇ):</strong><br>
            {finales['entalpia_h2_kj_kg']:.2f} kJ/kg
        </div>
        """, unsafe_allow_html=True
    )

st.markdown("---")
# --- Pesta√±as con el detalle del c√°lculo ---
tab2, = st.tabs(["Detalle del Proceso"])

with tab2:
    st.subheader("Condiciones y Propiedades")
    st.markdown("**Datos de Entrada Seleccionados:**")
    st.json(entradas)

    st.markdown("**Propiedades del Agua a la Entrada (L√≠quido Saturado):**")
    st.json({
        "volumen_especifico_m3_kg": f"{props['volumen_especifico_m3_kg']:.6f}",
        "entalpia_h1_kj_kg": f"{props['entalpia_h1_kj_kg']:.2f}"
    })

st.markdown("---")
st.caption("Developed by Juan Fernando Montoya Ortiz -Electrical Engineering Student -Universidad Nacional de Colombia")
```

```python
%%writefile B-Caldera.py

import streamlit as st
from iapws import IAPWS97
import math
from typing import Dict, Any

# Se agregan las clases .resultado-final y .nota-info para los nuevos colores.
custom_css = """

<style>
    .stApp {
        background-color: #0e1a40;
        color: #E0E0E0;
        font-family: 'Courier New', monospace;
    }

    h1, h2, h3 {
        color: #00BFFF;
    }

    .main .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
        background-color: #1B1D2B;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] {
        background-color: #222f5b;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] * {
        color: #c6e2ff !important;
    }

    .resultado-final {
        color: #FFD700;
        background-color: #2c3e50;
        border: 1px solid #FFD700;
        border-radius: 8px;
        padding: 1rem;
        text-align: center;
        font-size: 1.1rem;
    }

    .nota-info {
        color: #98FB98;
    }



    /* Estilo para el texto de la nota */
    .nota-info {
        color: #98FB98; /* Color Verde P√°lido para las letras */
    }
</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)


# --- MODELO (L√≥gica de c√°lculo) ---
@st.cache_data
def calcular_propiedades_caldera(
    P_in: float,
    T_in_C: float,
    P_out: float,
    T_out_C: float,
    V_out: float,
    D_mm: float
) -> Dict[str, Any]:
    """
    Calcula las propiedades termodin√°micas en una caldera.
    """
    try:
        D_m = D_mm / 1000
        A = (math.pi * D_m**2) / 4
        T_in_K = T_in_C + 273.15
        T_out_K = T_out_C + 273.15

        estado_salida = IAPWS97(P=P_out, T=T_out_K)
        v_out = estado_salida.v

        m_dot = (A * V_out) / v_out

        estado_entrada = IAPWS97(P=P_in, T=T_in_K)
        v_in = estado_entrada.v

        V_dot_in = m_dot * v_in
        V_in = V_dot_in / A

        resultados = {
            "area_m2": A,
            "v_especifico_entrada_m3_kg": v_in,
            "v_especifico_salida_m3_kg": v_out,
            "flujo_masico_kg_s": m_dot,
            "flujo_volumetrico_entrada_m3_s": V_dot_in,
            "velocidad_entrada_m_s": V_in,
            "error": None
        }
    except Exception as e:
        resultados = {"error": f"No se pudieron calcular las propiedades. Verifique los valores. Error: {e}"}

    return resultados


# =========================
# üöÄ Interfaz Streamlit
# =========================
col_img, col_title = st.columns([0.2, 1])
with col_img:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/%C3%8Dcono%20de%20calderas%20minimalista%20y%20funcional.png", width=200)
with col_title:
    st.title("Caldera: An√°lisis Interactivo del Comportamiento T√©rmico")
st.markdown("#### Explora c√≥mo var√≠an la temperatura, la presi√≥n y la entalp√≠a del fluido en una caldera de vapor.")
st.write("---")



# =========================
# üìò Teor√≠a
# =========================
with st.expander("Fundamentos Termodin√°micos de Calderas (1¬™ Ley ‚Äì Sistemas Abiertos)"):
    st.markdown(r"""
Las **calderas** son dispositivos termodin√°micos dise√±ados para **transferir calor a un fluido**, generalmente agua, con el fin de **generar vapor** que se utilizar√° en procesos industriales o generaci√≥n de energ√≠a.

Desde el punto de vista termodin√°mico, una caldera es un **sistema abierto en r√©gimen estacionario**, ya que existe flujo de masa (agua/vapor) a trav√©s del volumen de control.

---

### Aplicaci√≥n de la Primera Ley de la Termodin√°mica

Para un sistema abierto en estado estacionario, la **Primera Ley** se expresa como:

$$
\dot{Q} - \dot{W}_s = \dot{m} \left( h_{\text{salida}} - h_{\text{entrada}} + \frac{V_2^2 - V_1^2}{2} + g(z_2 - z_1) \right)
$$

Donde:
- $\dot{Q}$: Calor transferido al fluido [kW]
- $\dot{W}_s$: Trabajo de eje (en calderas t√≠picamente $\dot{W}_s = 0$)
- $\dot{m}$: Flujo m√°sico del fluido [kg/s]
- $h$: Entalp√≠a espec√≠fica [kJ/kg]
- $V$: Velocidad [m/s]
- $z$: Altura [m]
- $g$: Gravedad [9.81 m/s¬≤]

---

### Simplificaci√≥n para Calderas

En las calderas comunes:
- No hay trabajo de eje: $\dot{W}_s = 0$
- Cambios de energ√≠a cin√©tica y potencial son despreciables:
  $\frac{V_2^2 - V_1^2}{2} \approx 0$, $g(z_2 - z_1) \approx 0$

Entonces, la ecuaci√≥n se reduce a:

$$
\dot{Q} = \dot{m} (h_{\text{vapor}} - h_{\text{agua}})
$$

Esto implica que la **tasa de transferencia de calor** est√° directamente relacionada con el **salto ent√°lpico** entre el estado inicial (agua l√≠quida) y el estado final (vapor).

---

### Eficiencia t√©rmica de la caldera

Aunque las calderas no producen trabajo mec√°nico directo, se puede definir una eficiencia basada en el uso del combustible:

$$
\eta = \frac{\dot{m} (h_{\text{vapor}} - h_{\text{agua}})}{\dot{m}_{\text{comb}} \cdot \text{PCI}}
$$

Donde:
- $\dot{m}_{\text{comb}}$: Flujo m√°sico del combustible
- $\text{PCI}$: Poder calor√≠fico inferior del combustible

---

### Conclusi√≥n

Las calderas son un claro ejemplo de aplicaci√≥n de la **Primera Ley de la Termodin√°mica** en sistemas abiertos. Su an√°lisis permite determinar el **calor requerido** para generar vapor a partir de agua l√≠quida, esencial para el dise√±o y evaluaci√≥n energ√©tica de ciclos t√©rmicos como el de Rankine.

üìö **Fuente**: Adaptado de [termodinamica-1aa131.blogspot.com](https://termodinamica-1aa131.blogspot.com/2013/06/caldera.html)
""")

# =========================
# üß™ Ejercicio y Figura
# =========================
col1, col2 = st.columns(2)
with col1:
    with st.expander("üß™ Ejercicio Propuesto ‚Äì Caldera"):
        st.markdown("""
**Primera ley de la termodin√°mica ‚Äì Sistemas Abiertos**

**CALDERA**

Entra **Agua** a los tubos de una Caldera de **120 mm de di√°metro constante**, con una **Presi√≥n de 5 MPa** y una **Temperatura de 60¬∞C**, y **Sale** a una **Temperatura de 450¬∞C** y una **velocidad de 80 m/s**.

Determinar:

- a) El **flujo volum√©trico** de entrada.
- b) La **velocidad del Agua** a la entrada de este dispositivo.
---

**Fuente:** *Ejercicio tomado y adaptado de **LaMejorAsesor√≠aEducativa ‚Äì YouTube***.
""")
with col2:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Caldera.png",
             caption="**Comparaci√≥n esquem√°tica del funcionamiento interno de una caldera tipo locomotora.\\n\\nFuente: Adaptado de https://termodinamica-1aa131.blogspot.com/2013/06/caldera.html.")
    st.markdown("### Desarrollo visual")
    st.video("https://youtu.be/CuOPVWsVw5Q?si=VHYeOAiwX5XSV2kJ")
st.markdown("---")

# --- Barra Lateral con Sliders para Entradas del Usuario ---
st.sidebar.header("Par√°metros de Entrada")

p_in = st.sidebar.slider("Presi√≥n del Proceso (MPa)", min_value=0.1, max_value=25.0, value=5.0, step=0.1, help="La presi√≥n de entrada y salida se asumen iguales.")
t_in = st.sidebar.slider("Temp. Entrada (¬∞C)", min_value=10.0, max_value=1300.0, value=60.0, step=1.0)
t_out = st.sidebar.slider("Temp. Salida (¬∞C)", min_value=100.0, max_value=1300.0, value=450.0, step=1.0)
v_out = st.sidebar.slider("Velocidad Salida (m/s)", min_value=1.0, max_value=150.0, value=80.0, step=1.0)
d_mm = st.sidebar.slider("Di√°metro Tubo (mm)", min_value=10.0, max_value=500.0, value=120.0, step=1.0)

# --- L√≥gica Principal: Se asume P_out = P_in ---
# La presi√≥n de salida es igual a la de entrada para el caso ideal
p_out = p_in
resultados = calcular_propiedades_caldera(p_in, t_in, p_out, t_out, v_out, d_mm)

if resultados.get("error"):
    st.error(f"**Error de C√°lculo:** {resultados['error']}")
    st.warning("Algunas combinaciones de presi√≥n y temperatura no son f√≠sicamente posibles. Intenta ajustar los valores.")
else:
    st.header("üìä Resultados del An√°lisis")
    st.markdown("---")

    st.subheader("üéØ Objetivos Principales")
    col_res1, col_res2, col_res3 = st.columns(3)

    with col_res1:
        st.markdown(
            f"""
            <div class='resultado-final'>
                <strong>Flujo M√°sico (·πÅ):</strong><br>
                {resultados['flujo_masico_kg_s']:.2f} kg/s
            </div>
            """, unsafe_allow_html=True
        )

    with col_res2:
        st.markdown(
            f"""
            <div class='resultado-final'>
                <strong>Velocidad de Entrada (V‚ÇÅ):</strong><br>
                {resultados['velocidad_entrada_m_s']:.2f} m/s
            </div>
            """, unsafe_allow_html=True
        )

    with col_res3:
        st.markdown(
            f"""
            <div class='resultado-final'>
                <strong>Flujo Volum√©trico Entrada (VÃá):</strong><br>
                {resultados['flujo_volumetrico_entrada_m3_s']:.4f} m¬≥/s
            </div>
            """, unsafe_allow_html=True
        )

    st.markdown("---")
    st.subheader("üìò Propiedades Calculadas")

    info_col1, info_col2 = st.columns(2)

    with info_col1:
        st.markdown(
            f"""
            <div class='resultado-final'>
                <strong>√Årea del Tubo (A):</strong><br>
                {resultados['area_m2']:.5f} m¬≤
            </div>
            """, unsafe_allow_html=True
        )
        st.markdown(
            f"""
            <div class='resultado-final'>
                <strong>Vol. Espec√≠fico Entrada (v‚ÇÅ):</strong><br>
                {resultados['v_especifico_entrada_m3_kg']:.7f} m¬≥/kg
            </div>
            """, unsafe_allow_html=True
        )

    with info_col2:
        st.markdown(
            f"""
            <div class='resultado-final'>
                <strong>Vol. Espec√≠fico Salida (v‚ÇÇ):</strong><br>
                {resultados['v_especifico_salida_m3_kg']:.5f} m¬≥/kg
            </div>
            """, unsafe_allow_html=True
        )


st.markdown("---")
st.caption("Developed by Juan Fernando Montoya Ortiz -Electrical Engineering Student -Universidad Nacional de Colombia")
```

```python
%%writefile C-Turbinas.py



import streamlit as st
from iapws import IAPWS97
from typing import Dict, Any


custom_css = """

<style>
    .stApp {
        background-color: #0e1a40;
        color: #E0E0E0;
        font-family: 'Courier New', monospace;
    }

    h1, h2, h3 {
        color: #00BFFF;
    }

    .main .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
        background-color: #1B1D2B;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] {
        background-color: #222f5b;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] * {
        color: #c6e2ff !important;
    }

    .resultado-final {
        color: #FFD700;
        background-color: #2c3e50;
        border: 1px solid #FFD700;
        border-radius: 8px;
        padding: 1rem;
        text-align: center;
        font-size: 1.1rem;
    }

    .nota-info {
        color: #98FB98;
    }



    /* Estilo para el texto de la nota */
    .nota-info {
        color: #98FB98; /* Color Verde P√°lido para las letras */
    }
</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)
# ==============================================================================
# 1. FUNCI√ìN DE C√ÅLCULO PRINCIPAL
# Encapsula toda la l√≥gica del script original.
# ==============================================================================
def analizar_turbina(
    m_dot: float, P1_mpa: float, T1_c: float, V1: float,
    P2_kpa: float, x2: float, V2: float,
    q_loss_I: float, q_loss_II: float
) -> Dict[str, Any]:
    """
    Realiza el an√°lisis termodin√°mico completo de la turbina.

    Args:
        m_dot (float): Flujo m√°sico (kg/s).
        P1_mpa (float): Presi√≥n de entrada (MPa).
        T1_c (float): Temperatura de entrada (¬∞C).
        V1 (float): Velocidad de entrada (m/s).
        P2_kpa (float): Presi√≥n de salida (kPa).
        x2 (float): Calidad de vapor a la salida.
        V2 (float): Velocidad de salida (m/s).
        q_loss_I (float): P√©rdida de calor espec√≠fica para el caso I (kJ/kg).
        q_loss_II (float): P√©rdida de calor espec√≠fica para el caso II (kJ/kg).

    Returns:
        Dict[str, Any]: Un diccionario con todos los resultados calculados.
    """
    resultados = {}
    try:
        # --- Conversi√≥n de unidades ---
        P2_mpa = P2_kpa / 1000.0
        T1_k = T1_c + 273.15

        # --- a) Cambio de energ√≠a cin√©tica espec√≠fica (Œîec) ---
        delta_ec = (V2**2 - V1**2) / 2000.0

        # --- b) Propiedades termodin√°micas y √°reas ---
        state1 = IAPWS97(P=P1_mpa, T=T1_k)
        h1 = state1.h
        v1 = state1.v
        A1 = m_dot * v1 / V1

        state2 = IAPWS97(P=P2_mpa, x=x2)
        h2 = state2.h
        v2 = state2.v
        A2 = m_dot * v2 / V2

        delta_h = h1 - h2

        # --- c) C√°lculo de la Potencia (Wt) para las 4 condiciones ---
        Q_dot_I = -q_loss_I * m_dot
        Wt_I = m_dot * (delta_h - delta_ec) + Q_dot_I

        Q_dot_II = -q_loss_II * m_dot
        Wt_II = m_dot * delta_h + Q_dot_II

        Wt_III = m_dot * (delta_h - delta_ec)

        Wt_IV = m_dot * delta_h

        # Llenar el diccionario de resultados
        resultados = {
            "delta_ec": delta_ec, "h1": h1, "v1": v1, "A1": A1,
            "h2": h2, "v2": v2, "A2": A2, "delta_h": delta_h,
            "Wt_I": Wt_I, "Wt_II": Wt_II, "Wt_III": Wt_III, "Wt_IV": Wt_IV,
            "error": None
        }

    except Exception as e:
        resultados["error"] = str(e)

    return resultados

# ==============================================================================
# 2. INTERFAZ GR√ÅFICA CON STREAMLIT
# ==============================================================================

# --- Configuraci√≥n de la p√°gina ---
col_img, col_title = st.columns([0.2, 1])
with col_img:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Turbina%20y%20flecha%20en%20contraste.png", width=200)
with col_title:
    st.title("Turbina: An√°lisis Interactivo de Expansi√≥n de Vapor")
st.markdown("#### Analiza c√≥mo se transforma la energ√≠a del fluido en trabajo √∫til a trav√©s de una turbina.")
st.write("---")



# =========================
# üìò Teor√≠a
# =========================
with st.expander("üìò Fundamentos Termodin√°micos de Turbinas (√áengel, 7¬™ ed.)"):
    st.markdown(r"""
Las **turbinas** son dispositivos que convierten la energ√≠a contenida en un fluido en **trabajo mec√°nico √∫til**. Se utilizan ampliamente en **centrales t√©rmicas y hidroel√©ctricas**, donde impulsan generadores el√©ctricos.

En una turbina, el fluido realiza trabajo al expandirse y empujar los **√°labes** conectados a un eje giratorio, provocando su rotaci√≥n. El dispositivo entonces produce trabajo al entorno ($\dot{W}_{\text{salida}} > 0$).

Desde el punto de vista energ√©tico, el an√°lisis ideal de una turbina considera que:
- No hay **transferencia significativa de calor** ($\dot{Q} \approx 0$), debido al buen aislamiento t√©rmico.
- No hay cambio apreciable de **energ√≠a potencial** ni de **energ√≠a cin√©tica** ($\Delta ep \approx 0$, $\Delta ec \approx 0$).

Por lo tanto, la **Primera Ley de la Termodin√°mica** (para flujo estable) se simplifica como:

$$
\dot{W}_{\text{turbina}} = \dot{m}(h_1 - h_2)
$$

donde:
- $\dot{W}_{\text{turbina}}$ es el **trabajo generado por unidad de tiempo**
- $\dot{m}$ es el **flujo m√°sico**
- $h_1$ y $h_2$ son las **entalp√≠as espec√≠ficas** de entrada y salida

Este modelo permite estimar el trabajo que se puede obtener a partir de la ca√≠da de entalp√≠a del fluido durante su paso por la turbina.

üìö **Fuente**: √áengel, Yunus A., *Termodin√°mica*, 7¬™ Edici√≥n, McGraw-Hill, Secci√≥n 5.5 "Turbinas", pp. 279‚Äì280.
""")


# =========================
# üß™ Ejercicio y Figura
# =========================
col1, col2 = st.columns(2)
with col1:
    with st.expander("üß™ Ejercicio Propuesto ‚Äì Turbinas"):
        st.markdown(r"""
**Primera ley de la termodin√°mica ‚Äì Sistemas Abiertos**

**TURBINA**

Por una **turbina** pasa un flujo m√°sico estable de **agua** igual a **15 kg/s**.

- En la **entrada**:
  - Temperatura: **350‚ÄØ¬∞C**
  - Presi√≥n: **5 MPa**
  - Velocidad: **70 m/s**

- En la **salida**:
  - Presi√≥n: **75 kPa**
  - Calidad: **90%**
  - Velocidad: **40 m/s**

Se pide calcular:

- a) El **cambio de energ√≠a cin√©tica**: $\Delta e_C$
- b) El **√°rea** en la entrada y salida de la turbina
- c) La **potencia generada por la turbina** bajo los siguientes casos:

  - c.1) La turbina pierde **10 kJ/kg** y se considera $\Delta e_C$
  - c.2) La turbina pierde **80 kJ/s** y no se considera $\Delta e_C$
  - c.3) La turbina es **adiab√°tica** y se considera $\Delta e_C$
  - c.4) La turbina es **adiab√°tica** y **no** se considera $\Delta e_C$

---

**Fuente:** *Ejercicio tomado y adaptado de **LaMejorAsesor√≠aEducativa ‚Äì YouTube***.
""")
with col2:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/TurbinaBook.png",
             caption="**FIGURA 5-25** ‚Äì FIGURA 5-28** ‚Äì Esquema de turbina.\\n\\nFuente: √áengel ‚Äì Termodin√°mica, 7¬™ Edici√≥n.")
    st.markdown("### Desarrollo visual")
    st.video("https://youtu.be/_n2ozXyNBSc?si=5XJVptZt_PPEl59Y")

st.markdown("---")

# --- Barra lateral para los controles de entrada ---
st.sidebar.header("Par√°metros de Entrada")

# --- Condiciones Generales ---
m_dot = st.sidebar.number_input("Flujo M√°sico (·πÅ) [kg/s]", min_value=1.0, value=15.0, step=1.0)

# --- Condiciones de Entrada (Estado 1) ---
st.sidebar.subheader("Estado 1 (Entrada)")
P1_mpa = st.sidebar.slider("Presi√≥n (P‚ÇÅ) [MPa]", 1.0, 15.0, 5.0, 0.5)
T1_c = st.sidebar.slider("Temperatura (T‚ÇÅ) [¬∞C]", 200.0, 600.0, 350.0, 5.0)
V1 = st.sidebar.slider("Velocidad (V‚ÇÅ) [m/s]", 10, 200, 70, 5)

# --- Condiciones de Salida (Estado 2) ---
st.sidebar.subheader("Estado 2 (Salida)")
P2_kpa = st.sidebar.slider("Presi√≥n (P‚ÇÇ) [kPa]", 10.0, 200.0, 75.0, 5.0)
x2 = st.sidebar.slider("Calidad de Vapor (x‚ÇÇ)", 0.0, 1.0, 0.9, 0.01)
V2 = st.sidebar.slider("Velocidad (V‚ÇÇ) [m/s]", 10, 200, 40, 5)

# --- P√©rdidas de Calor ---
st.sidebar.subheader("P√©rdidas de Calor")
q_loss_I = st.sidebar.number_input("P√©rdida Caso I (q) [kJ/kg]", min_value=0.0, value=10.0, step=5.0)
q_loss_II = st.sidebar.number_input("P√©rdida Caso II (q) [kJ/kg]", min_value=0.0, value=80.0, step=5.0)

# --- Ejecuci√≥n y Presentaci√≥n de Resultados ---
resultados = analizar_turbina(m_dot, P1_mpa, T1_c, V1, P2_kpa, x2, V2, q_loss_I, q_loss_II)

st.divider()

if resultados.get("error"):
    st.error(f"**Error en el c√°lculo:** No se pudieron determinar las propiedades. Verifica los par√°metros. \n\n*Detalle: {resultados['error']}*")
else:
    # --- Columnas para mostrar resultados de forma organizada ---
    # --- Columnas para mostrar resultados de forma organizada ---
  col1, col2 = st.columns(2)

  with col1:
      st.subheader("Propiedades en la Entrada (1)")
      st.markdown(
          f"""
          <div class='resultado-final'>
              <strong>Entalp√≠a (h‚ÇÅ):</strong><br>
              {resultados['h1']:.2f} kJ/kg
          </div>
          <div class='resultado-final'>
              <strong>√Årea (A‚ÇÅ):</strong><br>
              {resultados['A1']:.4f} m¬≤
          </div>
          <div class='resultado-final'>
              <strong>Vol. Espec√≠fico (v‚ÇÅ):</strong><br>
              {resultados['v1']:.4f} m¬≥/kg
          </div>
          """, unsafe_allow_html=True
      )

  with col2:
      st.subheader("Propiedades en la Salida (2)")
      st.markdown(
          f"""
          <div class='resultado-final'>
              <strong>Entalp√≠a (h‚ÇÇ):</strong><br>
              {resultados['h2']:.2f} kJ/kg
          </div>
          <div class='resultado-final'>
              <strong>√Årea (A‚ÇÇ):</strong><br>
              {resultados['A2']:.4f} m¬≤
          </div>
          <div class='resultado-final'>
              <strong>Vol. Espec√≠fico (v‚ÇÇ):</strong><br>
              {resultados['v2']:.4f} m¬≥/kg
          </div>
          """, unsafe_allow_html=True
      )

  st.divider()

  st.subheader("Cambios de Energ√≠a")
  c1, c2 = st.columns(2)

  with c1:
      st.markdown(
          f"""
          <div class='resultado-final'>
              <strong>Œîh (h‚ÇÅ - h‚ÇÇ):</strong><br>
              {resultados['delta_h']:.2f} kJ/kg
          </div>
          """, unsafe_allow_html=True
      )
  with c2:
      st.markdown(
          f"""
          <div class='resultado-final'>
              <strong>Œîec:</strong><br>
              {resultados['delta_ec']:.2f} kJ/kg
          </div>
          """, unsafe_allow_html=True
      )

  st.divider()

  # --- Resultados de Potencia ---
  st.header("‚ö° Potencia Generada por la Turbina (·∫Üt)")
  st.markdown("Resultados para los cuatro escenarios de an√°lisis:")

  r1, r2 = st.columns(2)
  with r1:
      st.markdown(
          f"""
          <div class='resultado-final'>
              <strong>I. Con P√©rdida de {q_loss_I} kJ/kg y ŒîEc</strong><br>
              Potencia (·∫Üt): {resultados['Wt_I']:.2f} kW
          </div>
          <div class='resultado-final'>
              <strong>III. Proceso Adiab√°tico (Q=0) y con ŒîEc</strong><br>
              Potencia (·∫Üt): {resultados['Wt_III']:.2f} kW
          </div>
          """, unsafe_allow_html=True
      )

  with r2:
      st.markdown(
          f"""
          <div class='resultado-final'>
              <strong>II. Con P√©rdida de {q_loss_II} kJ/kg y sin ŒîEc</strong><br>
              Potencia (·∫Üt): {resultados['Wt_II']:.2f} kW
          </div>
          <div class='resultado-final'>
              <strong>IV. Proceso Adiab√°tico (Q=0) y sin ŒîEc</strong><br>
              Potencia (·∫Üt): {resultados['Wt_IV']:.2f} kW
          </div>
          """, unsafe_allow_html=True
      )

st.markdown("---")
st.caption("Developed by Juan Fernando Montoya Ortiz -Electrical Engineering Student -Universidad Nacional de Colombia")
```

```python
%%writefile D-compresor.py


### C√≥digo de la Aplicaci√≥n Streamlit (`app.py`)

import streamlit as st
import math





custom_css = """

<style>
    .stApp {
        background-color: #0e1a40;
        color: #E0E0E0;
        font-family: 'Courier New', monospace;
    }

    h1, h2, h3 {
        color: #00BFFF;
    }

    .main .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
        background-color: #1B1D2B;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] {
        background-color: #222f5b;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] * {
        color: #c6e2ff !important;
    }

    .resultado-final {
        color: #FFD700;
        background-color: #2c3e50;
        border: 1px solid #FFD700;
        border-radius: 8px;
        padding: 1rem;
        text-align: center;
        font-size: 1.1rem;
    }

    .nota-info {
        color: #98FB98;
    }


    /* Estilo para el texto de la nota */
    .nota-info {
        color: #98FB98; /* Color Verde P√°lido para las letras */
    }


</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)


# --- CONFIGURACI√ìN DE LA P√ÅGINA ---
st.set_page_config(
    page_title="An√°lisis de Compresor",
    page_icon="‚öôÔ∏è",
    layout="centered"
)

# --- T√çTULO Y DESCRIPCI√ìN ---
col_img, col_title = st.columns([0.2, 1])
with col_img:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Compresor%20de%20aire%20ilustre.png", width=200)
with col_title:
    st.title("Compresor: An√°lisis Interactivo del Trabajo de Compresi√≥n")
st.markdown("#### Explora c√≥mo var√≠an la presi√≥n, la temperatura y la entalp√≠a del gas durante el proceso de compresi√≥n.")


# =========================
# üìò Teor√≠a
# =========================
with st.expander("üìò Fundamentos Termodin√°micos de Compresores (√áengel, 7¬™ ed.)"):
    st.markdown(r"""
Los **compresores** son dispositivos utilizados para **aumentar la presi√≥n de un gas** mediante el suministro de **trabajo mec√°nico** desde una fuente externa. Son fundamentales en sistemas de refrigeraci√≥n, aire acondicionado, motores de combusti√≥n y procesos industriales.

En un compresor, el fluido (generalmente un gas) recibe energ√≠a en forma de **trabajo de eje**. A diferencia de las turbinas, los compresores **no generan trabajo**, sino que lo **consumen**:

$$
\dot{W}_{\text{entrada}} > 0
$$

Los compresores se analizan generalmente bajo estas suposiciones ideales:
- **Sin transferencia de calor significativa** ($\dot{Q} \approx 0$) ‚Üí proceso adiab√°tico
- **Sin cambios importantes en energ√≠a potencial** ($\Delta ep \approx 0$)
- **Cambios en energ√≠a cin√©tica despreciables**, salvo que el gas se acelere notablemente ($\Delta ec \approx 0$)

Aplicando la **Primera Ley de la Termodin√°mica** para flujo estacionario:

$$
\dot{W}_{\text{compresor}} = \dot{m} \left( h_2 - h_1 + \frac{V_2^2 - V_1^2}{2} + g(z_2 - z_1) \right)
$$

Y si se considera un compresor **adiab√°tico y sin variaciones de energ√≠a cin√©tica ni potencial**, se simplifica a:

$$
\dot{W}_{\text{compresor}} = \dot{m}(h_2 - h_1)
$$

donde:
- $h$ es la entalp√≠a espec√≠fica
- $\dot{m}$ es el flujo m√°sico
- $\dot{W}_{\text{compresor}}$ es el trabajo requerido por unidad de tiempo

Este modelo permite calcular el **trabajo de compresi√≥n** requerido para lograr un aumento de presi√≥n y temperatura en el gas.

üìö **Fuente**: √áengel, Yunus A., *Termodin√°mica*, 7¬™ Edici√≥n, McGraw-Hill, Secci√≥n 5.5 "Compresores", pp. 279‚Äì280.
""")

col1, col2 = st.columns(2)
with col1:
    with st.expander("üß™ Ejercicio Propuesto ‚Äì Compresor"):
        st.markdown("""
**Primera ley de la termodin√°mica ‚Äì Sistemas Abiertos**

**COMPRESOR**

Un compresor recibe **aire** a:
- **Presi√≥n:** 100 kPa
- **Temperatura:** 300 K
- **Velocidad:** 5 m/s

Y lo entrega a:
- **Presi√≥n:** 800 kPa
- **Temperatura:** 500 K
- **Velocidad:** 7 m/s

El **flujo m√°sico** de aire es de **0.3 kg/s**.
El **calor disipado** hacia el ambiente es **15 kJ/kg**.

Asuma que el aire se comporta como un **gas ideal**.

Determinar:
- a) El **flujo volum√©trico** a la entrada del compresor.
- b) La **potencia requerida** para operar el compresor.
- c) El **trabajo por unidad de masa**.
---
""")

with col2:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Compresor.png",
             caption="**FIGURA 5-27 - Esquema de compresor\\n\\nFuente: √áengel ‚Äì Termodin√°mica, 7¬™ Edici√≥n.")
    st.markdown("### Ejercicio adicional")
    st.video("https://www.youtube.com/watch?v=atcJEUzwVRk")
st.markdown("---")

# --- BARRA LATERAL CON DATOS DE ENTRADA ---
st.sidebar.header("Datos de Entrada")

# Agrupar condiciones de entrada y salida
with st.sidebar.expander("**Condiciones del Fluido**", expanded=True):
    m_dot = st.slider("Flujo m√°sico (m_dot) [kg/s]", min_value=0.0, max_value=2.0, value=0.3, step=0.01)
    q_out = st.slider("Calor disipado (q_out) [kJ/kg]", min_value=0.0, max_value=100.0, value=15.0, step=0.5)

with st.sidebar.expander("**Propiedades del Gas Ideal)**", expanded=True):
    cp = st.slider("Calor espec√≠fico (cp) [kJ/(kg¬∑K)]", min_value=0.9, max_value=1.2, value=1.005, step=0.001)
    R = st.slider("Constante del gas (R) [kJ/(kg¬∑K)]", min_value=0.25, max_value=0.35, value=0.287, step=0.001)

with st.sidebar.expander("**Condiciones de Entrada (1)**", expanded=True):
    P1 = st.slider("Presi√≥n 1 (P1) [kPa]", min_value=50.0, max_value=500.0, value=100.0, step=1.0)
    T1 = st.slider("Temperatura 1 (T1) [K]", min_value=250.0, max_value=600.0, value=300.0, step=1.0)
    V1 = st.slider("Velocidad 1 (V1) [m/s]", min_value=0.0, max_value=100.0, value=5.0, step=0.5)

with st.sidebar.expander("**Condiciones de Salida (2)**", expanded=True):
    P2 = st.slider("Presi√≥n 2 (P2) [kPa]", min_value=100.0, max_value=1500.0, value=800.0, step=5.0)
    T2 = st.slider("Temperatura 2 (T2) [K]", min_value=300.0, max_value=1000.0, value=500.0, step=1.0)
    V2 = st.slider("Velocidad 2 (V2) [m/s]", min_value=0.0, max_value=100.0, value=7.0, step=0.5)


# --- C√ÅLCULOS TERMODIN√ÅMICOS ---

# a) C√°lculo del flujo volum√©trico de entrada (V_dot_1)
if P1 == 0:
    V_dot_1 = 0
    st.error("Error: La presi√≥n de entrada P1 no puede ser cero.")
else:
    V_dot_1 = (m_dot * R * T1) / P1

# b) C√°lculo de la potencia requerida por el compresor (W_dot)
# 1. Tasa de transferencia de calor (Q_dot). Negativa porque sale del sistema.
Q_dot = -m_dot * q_out

# 2. Cambio de entalp√≠a espec√≠fica (delta_h).
delta_h = cp * (T2 - T1)

# 3. Cambio de energ√≠a cin√©tica espec√≠fica (delta_ec). Se convierte de J/kg a kJ/kg.
delta_ec = (V2**2 - V1**2) / (2 * 1000)

# 4. Potencia (W_dot). El resultado es negativo, indicando que se consume potencia.
W_dot = Q_dot - m_dot * (delta_h + delta_ec)
potencia_requerida = abs(W_dot)

# c) C√°lculo del trabajo por unidad de masa (w)
if m_dot == 0:
    w = 0
    st.error("Error: El flujo m√°sico (m_dot) no puede ser cero para calcular el trabajo por masa.")
else:
    w = W_dot / m_dot

# --- PRESENTACI√ìN DE RESULTADOS ---

st.header("Resultados del An√°lisis")

# M√©tricas principales
col1, col2, col3 = st.columns(3)

with col1:
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Potencia Requerida:</strong><br>
            {potencia_requerida:.3f} kW
        </div>
        """, unsafe_allow_html=True
    )

with col2:
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Flujo Volum√©trico:</strong><br>
            {V_dot_1:.4f} m¬≥/s
        </div>
        """, unsafe_allow_html=True
    )

with col3:
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Trabajo por Masa:</strong><br>
            {w:.3f} kJ/kg
        </div>
        """, unsafe_allow_html=True
    )

# Desglose de los c√°lculos
with st.expander("Ver desglose de los c√°lculos", expanded=False):
    st.subheader("b) Desglose del C√°lculo de Potencia")

    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Tasa de calor disipado (QÃá):</strong><br>
            {Q_dot:.3f} kW
        </div>
        """, unsafe_allow_html=True
    )

    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Cambio de entalp√≠a espec√≠fica (h‚ÇÇ - h‚ÇÅ):</strong><br>
            {delta_h:.3f} kJ/kg
        </div>
        """, unsafe_allow_html=True
    )

    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Cambio de energ√≠a cin√©tica espec√≠fica (Œîec):</strong><br>
            {delta_ec:.4f} kJ/kg
        </div>
        """, unsafe_allow_html=True
    )

    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Potencia de entrada al compresor (·∫ÜÃá):</strong><br>
            {W_dot:.3f} kW
        </div>
        <p><em>El signo negativo indica que el trabajo es una entrada de energ√≠a al sistema (consumo).</em></p>
        """, unsafe_allow_html=True
    )

# --- SECCI√ìN DE F√ìRMULAS ---
st.header("Fundamento Te√≥rico")
with st.expander("Ver F√≥rmulas Utilizadas", expanded=False):
    st.subheader("a) Flujo Volum√©trico (Ecuaci√≥n del Gas Ideal)")
    st.latex(r'''P_1 \dot{V}_1 = \dot{m} R T_1 \quad \implies \quad \dot{V}_1 = \frac{\dot{m} R T_1}{P_1}''')

    st.subheader("b) Potencia (Primera Ley de la Termodin√°mica)")
    st.write("Para un sistema abierto en estado estacionario, el balance de energ√≠a es:")
    st.latex(r'''\dot{Q} - \dot{W} = \dot{m} \left[ (h_2 - h_1) + \frac{V_2^2 - V_1^2}{2} + g(z_2 - z_1) \right]''')
    st.write("Asumiendo que el cambio de energ√≠a potencial es despreciable y despejando la potencia $\dot{W}$:")
    st.latex(r'''\dot{W} = \dot{Q} - \dot{m} \left[ \Delta h + \Delta e_c \right]''')
    st.write("Donde:")
    st.latex(r'''\dot{Q} = -\dot{m} \cdot q_{out}''')
    st.latex(r'''\Delta h = c_p (T_2 - T_1)''')
    st.latex(r'''\Delta e_c = \frac{V_2^2 - V_1^2}{2000} \quad (\text{en kJ/kg})''')

    st.subheader("c) Trabajo por Unidad de Masa")
    st.latex(r'''w = \frac{\dot{W}}{\dot{m}}''')

st.markdown("---")
st.caption("Developed by Juan Fernando Montoya Ortiz -Electrical Engineering Student -Universidad Nacional de Colombia")
```

```python
%%writefile E-Tobera.py
# pages/Tobera.py ‚Äì Dashboard interactivo para an√°lisis de tobera

import streamlit as st
import numpy as np

custom_css = """

<style>
    .stApp {
        background-color: #0e1a40;
        color: #E0E0E0;
        font-family: 'Courier New', monospace;
    }

    h1, h2, h3 {
        color: #00BFFF;
    }

    .main .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
        background-color: #1B1D2B;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] {
        background-color: #222f5b;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] * {
        color: #c6e2ff !important;
    }

    .resultado-final {
        color: #FFD700;
        background-color: #2c3e50;
        border: 1px solid #FFD700;
        border-radius: 8px;
        padding: 1rem;
        text-align: center;
        font-size: 1.1rem;
    }

    .nota-info {
        color: #98FB98;
    }


    /* Estilo para el texto de la nota */
    .nota-info {
        color: #98FB98; /* Color Verde P√°lido para las letras */
    }


</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)


# =========================
# üìù Constantes globales
# =========================
R_kJ_kgK = 0.287   # kJ/kg¬∑K
Cp_kJ_kgK = 1.005  # kJ/kg¬∑K

# =========================
# üìù Funci√≥n de an√°lisis de tobera
# =========================

def analizar_tobera(P1_kPa, T1_C, V1_ms, A1_cm2, P2_kPa, V2_ms):
    P1_Pa, P2_Pa = P1_kPa * 1000, P2_kPa * 1000
    T1_K = T1_C + 273.15
    A1_m2 = A1_cm2 / 1e4

    v1 = (R_kJ_kgK * 1000 * T1_K) / P1_Pa
    m_dot = (A1_m2 * V1_ms) / v1

    Cp_J = Cp_kJ_kgK * 1000
    delta_T = (V1_ms**2 - V2_ms**2) / (2 * Cp_J)
    T2_K = T1_K + delta_T
    T2_C = T2_K - 273.15

    v2 = (R_kJ_kgK * 1000 * T2_K) / P2_Pa
    A2_m2 = (m_dot * v2) / V2_ms
    A2_cm2 = A2_m2 * 1e4

    return {
        "m_dot": m_dot,
        "T2_C": T2_C,
        "T2_K": T2_K,
        "A2_cm2": A2_cm2,
        "ratio": A2_m2 / A1_m2,
        "v1": v1,
        "v2": v2
    }

# =========================
# üöÄ Interfaz Streamlit
# =========================
col_img, col_title = st.columns([0.2, 1])
with col_img:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Toberas.png", width=200)
with col_title:
    st.title("Tobera: An√°lisis interactivo de Tobera de Aire")
st.markdown("#### Explora c√≥mo var√≠an la velocidad, la presi√≥n y la temperatura en una tobera.")
st.write("---")



# =========================
# üìò Teor√≠a
# =========================
with st.expander("üìò Fundamentos Termodin√°micos de Toberas y Difusores (√áengel, 7¬™ ed.)"):
    st.markdown(r"""
Las **toberas** y los **difusores** son dispositivos fundamentales en sistemas de propulsi√≥n como motores a chorro, cohetes, turbinas de gas y tambi√©n en aplicaciones industriales.
- Una **tobera** incrementa la **velocidad del fluido** a costa de una **disminuci√≥n de la presi√≥n**.
- Un **difusor**, en cambio, **disminuye la velocidad** del fluido mientras **aumenta su presi√≥n**.

Ambos dispositivos est√°n dise√±ados para modificar la energ√≠a cin√©tica del fluido que los atraviesa, y **no implican trabajo mec√°nico** ($\dot{W} \equiv 0$), ni transferencia de calor significativa ($\dot{Q} \approx 0$), y el cambio de energ√≠a potencial es despreciable ($\Delta ep \approx 0$).

Por ello, la **conservaci√≥n de energ√≠a** (1¬™ Ley de la Termodin√°mica para sistemas abiertos) se simplifica en estos casos a:

$$
h_1 + \frac{V_1^2}{2} = h_2 + \frac{V_2^2}{2}
$$

donde:
- $h = C_p T$ es la entalp√≠a espec√≠fica
- $V$ es la velocidad del flujo

Esto permite determinar la **temperatura de salida** $T_2$ si se conocen $T_1$, $V_1$ y $V_2$.

Asimismo, al aplicar la ecuaci√≥n de conservaci√≥n de masa:

$$
\dot{m} = \frac{A_1 V_1}{v_1} = \frac{A_2 V_2}{v_2}
$$

es posible calcular el **√°rea de salida** $A_2$, siendo:
- $v = \dfrac{RT}{P}$ el volumen espec√≠fico del aire como gas ideal.

üìö **Fuente**: √áengel, Yunus A., *Termodin√°mica*, 7¬™ Edici√≥n, McGraw-Hill, Secci√≥n 5.4 "Toberas y Difusores", pp. 278‚Äì279.
""")

# =========================
# üß™ Ejercicio y Figura
# =========================
col1, col2 = st.columns(2)
with col1:
    with st.expander("üß™ Ejercicio Propuesto ‚Äì Toberas"):
        st.markdown("""
**Primera ley de la termodin√°mica ‚Äì Sistemas Abiertos**

**TOBERAS**

La entrada de una tobera tiene una secci√≥n de 100 cm¬≤, y por ella entra aire a una velocidad de 50 m/s, con una presi√≥n de 350 kPa y una temperatura de 227 ¬∞C. Luego sale a una velocidad de 190 m/s y una presi√≥n de 120 kPa. Determinar:

- a) El flujo m√°sico de aire que circula por la tobera
- b) La temperatura del aire a la salida de la tobera
- c) El √°rea de secci√≥n de salida de la tobera
- d) La relaci√≥n de √°reas de entrada y salida

---

**Fuente:** *Ejercicio tomado y adaptado de **LaMejorAsesor√≠aEducativa ‚Äì YouTube***.
""")
with col2:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Imagen%201.png",
             caption="**FIGURA 5-25** ‚Äì La forma de toberas y difusores es tal que causan grandes cambios en la velocidad del fluido y, por lo tanto, en la energ√≠a cin√©tica.\\n\\nFuente: √áengel ‚Äì Termodin√°mica, 7¬™ Edici√≥n.")
    st.markdown("### Desarrollo visual")
    st.video("https://youtu.be/bK3nX37py_I?si=bWxFM11AvblVTWix")
st.markdown("---")

# Sidebar ‚Äì Par√°metros de entrada
tsidebar = st.sidebar
tsidebar.header("üì• Par√°metros de Entrada")

P1 = tsidebar.slider("Presi√≥n de entrada P‚ÇÅ [kPa]", min_value=50.0, max_value=1000.0, value=350.0, step=10.0)
T1 = tsidebar.slider("Temperatura de entrada T‚ÇÅ [¬∞C]", min_value=-50.0, max_value=800.0, value=227.0, step=5.0)
V1 = tsidebar.slider("Velocidad de entrada V‚ÇÅ [m/s]", min_value=1.0, max_value=500.0, value=50.0, step=1.0)
A1 = tsidebar.slider("√Årea de entrada A‚ÇÅ [cm¬≤]", min_value=1.0, max_value=1000.0, value=100.0, step=1.0)
P2 = tsidebar.slider("Presi√≥n de salida P‚ÇÇ [kPa]", min_value=5.0, max_value=1000.0, value=120.0, step=5.0)
V2 = tsidebar.slider("Velocidad de salida V‚ÇÇ [m/s]", min_value=1.0, max_value=1000.0, value=190.0, step=1.0)


if tsidebar.button("üí° Calcular"):
    res = analizar_tobera(P1, T1, V1, A1, P2, V2)

    st.subheader("üìä Resultados del An√°lisis")
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Flujo m√°sico ·πÅ:</strong> {res['m_dot']:.3f} kg/s
        </div>
        """, unsafe_allow_html=True
    )
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Temperatura de salida T‚ÇÇ:</strong> {res['T2_C']:.2f} ¬∞C ({res['T2_K']:.2f} K)
        </div>
        """, unsafe_allow_html=True
    )
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>√Årea de salida A‚ÇÇ:</strong> {res['A2_cm2']:.2f} cm¬≤
        </div>
        """, unsafe_allow_html=True
    )
    st.markdown(
        f"""
        <div class='resultado-final'>
            <strong>Relaci√≥n de √°reas A‚ÇÇ/A‚ÇÅ:</strong> {res['ratio']:.4f}
        </div>
        """, unsafe_allow_html=True
    )


    with st.expander("üîç Detalles de los C√°lculos"):
        st.write(
            f"""
            - **Volumen espec√≠fico entrada (v‚ÇÅ):** {res['v1']:.5f} m¬≥/kg
            - **Volumen espec√≠fico salida (v‚ÇÇ):** {res['v2']:.5f} m¬≥/kg
            - **Constante del gas R:** {R_kJ_kgK} kJ/kg¬∑K
            - **Calor espec√≠fico Cp:** {Cp_kJ_kgK} kJ/kg¬∑K
            """

        )

# =========================
# üë§ Cr√©ditos
# =========================
st.markdown("---")

st.caption("Developed by Juan Fernando Montoya Ortiz -Electrical Engineering Student -Universidad Nacional de Colombia")
```

```python
%%writefile F-Condensador.py
import streamlit as st
from iapws import IAPWS97
import sys

# Se agregan las clases .resultado-final y .nota-info para los nuevos colores.
custom_css = """

<style>
    .stApp {
        background-color: #0e1a40;
        color: #E0E0E0;
        font-family: 'Courier New', monospace;
    }

    h1, h2, h3 {
        color: #00BFFF;
    }

    .main .block-container {
        padding-top: 2rem;
        padding-bottom: 2rem;
        background-color: #1B1D2B;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] {
        background-color: #222f5b;
        border-radius: 10px;
    }

    section[data-testid="stSidebar"] * {
        color: #c6e2ff !important;
    }

    .resultado-final {
        color: #FFD700;
        background-color: #2c3e50;
        border: 1px solid #FFD700;
        border-radius: 8px;
        padding: 1rem;
        text-align: center;
        font-size: 1.1rem;
    }

    .nota-info {
        color: #98FB98;
    }



    /* Estilo para el texto de la nota */
    .nota-info {
        color: #98FB98; /* Color Verde P√°lido para las letras */
    }
</style>
"""
st.markdown(custom_css, unsafe_allow_html=True)





def calcular_flujo_enfriamiento(m_dot_v_kg_min, P_kPa, calidad_vapor, delta_T_agua_C):
    """
    Calcula el flujo m√°sico de agua de enfriamiento necesario en un condensador.

    Utiliza la Primera Ley de la Termodin√°mica para sistemas abiertos en estado
    estacionario. Las propiedades del vapor se obtienen con el est√°ndar IAPWS-IF97.

    Args:
        m_dot_v_kg_min (float): Flujo m√°sico del vapor de entrada en kg/min.
        P_kPa (float): Presi√≥n del vapor de entrada en kPa.
        calidad_vapor (float): Calidad del vapor de entrada (un valor entre 0 y 1).
        delta_T_agua_C (float): Aumento de temperatura permitido para el agua de
                                 enfriamiento en grados Celsius.

    Returns:
        dict: Un diccionario con todos los resultados de los c√°lculos.
        None: Si ocurre un error (ej. presi√≥n fuera de rango).
    """
    # --- Constantes ---
    # Calor espec√≠fico del agua l√≠quida en kJ/kg*¬∞C
    CP_AGUA = 4.18

    # --- Paso 1: Conversi√≥n de Unidades ---
    # Convertir el flujo m√°sico del vapor de kg/min a kg/s
    m_dot_v_kg_s = m_dot_v_kg_min / 60.0

    # --- Paso 2: Propiedades Termodin√°micas del Vapor ---
    # La librer√≠a iapws requiere la presi√≥n en MPa
    P_MPa = P_kPa / 1000.0

    # Crear objetos para el estado de l√≠quido y vapor saturado a la presi√≥n dada
    try:
        liq_saturado = IAPWS97(P=P_MPa, x=0)
        vap_saturado = IAPWS97(P=P_MPa, x=1)
    except ValueError as e:
        st.error(f"Error: La presi√≥n {P_kPa} kPa est√° fuera del rango v√°lido para el modelo IAPWS-97. {e}")
        return None
    except Exception as e:
        st.error(f"Ocurri√≥ un error inesperado al obtener propiedades del vapor: {e}")
        return None

    # Obtener entalp√≠as espec√≠ficas (h_f y h_g) en kJ/kg
    h_f = liq_saturado.h  # Entalp√≠a del l√≠quido saturado
    h_g = vap_saturado.h  # Entalp√≠a del vapor saturado
    h_fg = h_g - h_f      # Entalp√≠a de vaporizaci√≥n

    # --- Paso 3: C√°lculo de Entalp√≠as en Entrada y Salida ---
    # Entalp√≠a del vapor a la entrada (mezcla l√≠quido-vapor)
    h_entrada = h_f + calidad_vapor * h_fg

    # Entalp√≠a a la salida (l√≠quido saturado)
    h_salida = h_f

    # Diferencia de entalp√≠a del vapor
    delta_h_v = h_entrada - h_salida

    # --- Paso 4: Calor Liberado por el Vapor (Tasa de Transferencia) ---
    # Q_punto es la energ√≠a transferida por unidad de tiempo en kW
    Q_punto_kW = m_dot_v_kg_s * delta_h_v

    # --- Paso 5: C√°lculo del Flujo M√°sico del Agua de Enfriamiento ---
    # Q_punto = m_dot_a * cp_agua * Delta_T
    # Despejamos el flujo m√°sico del agua (m_dot_a)
    if delta_T_agua_C == 0:
        st.warning("El incremento de temperatura del agua no puede ser cero. Por favor, ingrese un valor mayor que cero.")
        return None
    m_dot_a_kg_s = Q_punto_kW / (CP_AGUA * delta_T_agua_C)

    # --- Empaquetar resultados ---
    resultados = {
        "m_dot_v_kg_s": m_dot_v_kg_s,
        "P_kPa": P_kPa,
        "calidad_vapor": calidad_vapor,
        "delta_T_agua_C": delta_T_agua_C,
        "h_f_kJ_kg": h_f,
        "h_g_kJ_kg": h_g,
        "h_fg_kJ_kg": h_fg,
        "h_entrada_kJ_kg": h_entrada,
        "h_salida_kJ_kg": h_salida,
        "Q_punto_kW": Q_punto_kW,
        "m_dot_a_kg_s": m_dot_a_kg_s
    }
    return resultados

# --- Dise√±o de la aplicaci√≥n Streamlit ---
st.set_page_config(
    page_title="C√°lculo de Condensador",
    page_icon="üí®",
    layout="wide"
)

# --- Barra Lateral (Sidebar) ---
st.sidebar.header("1. Datos de Entrada")

flujo_masico_vapor = st.sidebar.slider(
    "Flujo m√°sico de vapor (kg/min)",
    min_value=1.0, max_value=1000.0, value=420.0, step=1.0,
    help="Flujo m√°sico del vapor que entra al condensador."
)
presion_vapor = st.sidebar.slider(
    "Presi√≥n del vapor (kPa)",
    min_value=5.0, max_value=2000.0, value=30.0, step=0.1,
    help="Presi√≥n a la que se encuentra el vapor."
)
calidad_inicial = st.sidebar.slider(
    "Calidad del vapor (x)",
    min_value=0.0, max_value=1.0, value=0.90, step=0.01,
    help="Fracci√≥n de masa de vapor en la mezcla (0=l√≠quido saturado, 1=vapor saturado)."
)
delta_temperatura = st.sidebar.slider(
    "Incremento de temperatura del agua (¬∞C)",
    min_value=1.0, max_value=30.0, value=10.0, step=0.5,
    help="Aumento de temperatura permitido para el agua de enfriamiento."
)

# --- Contenido Principal ---
col_img, col_title = st.columns([0.2, 1])
with col_img:
    st.image("https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Intercambiador%20de%20calor%20moderno.png", width=200)
with col_title:
    st.title("Intercambiador de calor: An√°lisis interactivo de Intercambiador de calor")
st.markdown("""
**Este m√≥dulo interactivo** te permite analizar las condiciones de operaci√≥n de un intercambiador de calor de forma sencilla.
Podr√°s observar c√≥mo cambian variables clave como **presi√≥n**, **temperatura** y **velocidad** del fluido, y calcular par√°metros importantes como el **flujo m√°sico de enfriamiento necesario**.
""")

st.write("---")


with st.expander("üìò Fundamentos Termodin√°micos de Compresores (√áengel, 7¬™ ed.)"):
    st.markdown(r"""
        Los **compresores** son dispositivos mec√°nicos que elevan la presi√≥n de un gas al reducir su volumen. Son ampliamente utilizados en sistemas de refrigeraci√≥n, aire acondicionado, turbom√°quinas, y plantas industriales.

        A diferencia de las bombas (que trabajan con l√≠quidos), los compresores manipulan **gases** y generalmente requieren **trabajo de entrada** para operar, debido a la compresibilidad del fluido.

        **Principios Clave:**
        - **Primera Ley para sistemas abiertos:** Se analiza aplicando un balance de energ√≠a en r√©gimen estacionario.
        - **Trabajo de compresi√≥n:** El trabajo requerido por unidad de masa est√° dado por:
          $$
          w = h_2 - h_1
          $$
          donde \( h_1 \) y \( h_2 \) son las entalp√≠as del gas a la entrada y salida del compresor.
        - **Eficiencia isentr√≥pica:** Compara el trabajo real del compresor con el trabajo ideal (reversible y adiab√°tico).
        - **Aumento de temperatura y presi√≥n:** Durante la compresi√≥n, ambos par√°metros aumentan, dependiendo del tipo de proceso (isentr√≥pico, politr√≥pico o real).

        üìö **Fuente**: √áengel, Y. A., & Boles, M. A. (2011). *Termodin√°mica* (7¬™ ed.). McGraw-Hill.
    """)

col1, col2 = st.columns(2)
with col1:
    with st.expander("üß™ Ejercicio Propuesto ‚Äì Compresor"):
        st.markdown("""
Al **Condensador** de una Termoel√©ctrica entran **420 kg/min de Vapor** con una
**Presi√≥n de 30 kPa** y una **Calidad del 90%**; se requiere que el vapor
salga como **l√≠quido saturado**.

Para enfriar el vapor, se utiliza **agua de un r√≠o**, haci√©ndola pasar por el interior del condensador.
Para prevenir la contaminaci√≥n t√©rmica, el **agua del r√≠o** no debe superar un incremento
de temperatura igual a **10 ¬∞C**.

---

**Calcular:**

El **flujo m√°sico de agua de enfriamiento** necesario para satisfacer el cambio demandado por el vapor.

---

üìö **Fuente**: *Ejercicio tomado y adaptado de **LaMejorAsesor√≠aEducativa ‚Äì YouTube***.
        """)


with col2:
    st.image(
        "https://raw.githubusercontent.com/Jmontoyaor/thermodynamics/main/IMAGENES/Imagen%202.png",
        caption="Figura 5-35, Un intercambiador de calor puede ser tan simple como dos tuber√≠as conc√©ntricas. Fuente: √áengel ‚Äì Termodin√°mica, 7¬™ Edici√≥n.",
        use_container_width=True
    )

    st.markdown("### Desarrollo visual")
    st.video("https://youtu.be/XeOi0Rcqxio?si=Lv0RB7-jzc3riA-6")

st.markdown("---")

st.header("2. Resultados del C√°lculo")

if st.sidebar.button("Calcular Flujo de Enfriamiento", use_container_width=True):
    resultados_calculados = calcular_flujo_enfriamiento(
        m_dot_v_kg_min=flujo_masico_vapor,
        P_kPa=presion_vapor,
        calidad_vapor=calidad_inicial,
        delta_T_agua_C=delta_temperatura
    )

    if resultados_calculados:
        st.subheader("Datos de Entrada Utilizados:")
        st.write(f"- **Flujo m√°sico de vapor (·πÅ_v):** {flujo_masico_vapor:.1f} kg/min ({resultados_calculados['m_dot_v_kg_s']:.2f} kg/s)")
        st.write(f"- **Presi√≥n del vapor (P):** {presion_vapor} kPa")
        st.write(f"- **Calidad del vapor (x):** {calidad_inicial:.2f}")
        st.write(f"- **Incremento de temp. del agua (ŒîT):** {delta_temperatura} ¬∞C")

        st.subheader(f"Propiedades del Vapor (a {presion_vapor} kPa):")
        st.write(f"- **Entalp√≠a del l√≠quido saturado ($h_f$):** {resultados_calculados['h_f_kJ_kg']:.2f} kJ/kg")
        st.write(f"- **Entalp√≠a del vapor saturado ($h_g$):** {resultados_calculados['h_g_kJ_kg']:.2f} kJ/kg")
        st.write(f"- **Entalp√≠a de vaporizaci√≥n ($h_{{fg}}$):** {resultados_calculados['h_fg_kJ_kg']:.2f} kJ/kg")

        st.subheader("Balance de Energ√≠a para el Vapor:")
        st.write(f"- **Entalp√≠a de entrada ($h_{{entrada}}$):** {resultados_calculados['h_entrada_kJ_kg']:.2f} kJ/kg")
        st.write(f"- **Entalp√≠a de salida ($h_{{salida}}$):** {resultados_calculados['h_salida_kJ_kg']:.2f} kJ/kg")
        st.write(f"**- Tasa de calor liberado por el vapor ($\\dot{{Q}}$):** {resultados_calculados['Q_punto_kW']:.2f} kW")

        st.markdown("---")
        # Se reemplaza st.success con st.markdown para usar la clase CSS personalizada.
        st.markdown("<div class='resultado-final'>", unsafe_allow_html=True)

        st.latex( rf"""\textbf{{Flujo m√°sico de agua de enfriamiento necesario }} (\dot{{m}}_{{a}}) = {resultados_calculados['m_dot_a_kg_s']:.2f} \; \text{{kg/s}}""")

        st.markdown("</div>", unsafe_allow_html=True)

    else:
        st.warning("No se pudo realizar el c√°lculo. Por favor, revise los datos de entrada.")

st.markdown("---")
# Se reemplaza st.info con st.markdown para usar la clase CSS personalizada.
st.markdown("""
<div class='nota-info'>
<b>Nota sobre la precisi√≥n:</b> Este script utiliza el est√°ndar internacional IAPWS-IF97 para obtener las propiedades termodin√°micas del agua y el vapor, lo que garantiza una alta precisi√≥n en los c√°lculos. Los resultados pueden diferir ligeramente de los obtenidos con tablas de vapor antiguas.
</div>
""", unsafe_allow_html=True)


st.markdown("---")
st.caption("Developed by Juan Fernando Montoya Ortiz -Electrical Engineering Student -Universidad Nacional de Colombia")
```
